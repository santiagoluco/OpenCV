```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.imread('mild-steel-nominal-bore-tube.jpg',0)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
display_img(img,cmap='gray')
```

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
laplacian = cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian,cmap='gray')
```

```python
kernel = np.ones((4,4),np.uint8)
gradient = cv2.morphologyEx(sobelx,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```

```python
ret,th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY_INV)
display_img(th1,cmap='gray')
```

```python
gradient = cv2.morphologyEx(th1,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient,cmap='gray')
```

```python
imagem = cv2.bitwise_not(gradient)
```

```python
display_img(imagem,cmap='gray')
```

```python
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
plt.imshow(thresh5,cmap='gray')
```
